```python
import numpy as np
import pandas as pd
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import random 
from nltk.corpus import stopwords
import string
import nltk
import re
```

```python
fraud_df=pd.read_csv('./datasets/fraudulent_text_data.csv')
```

```python
fraud_df.frauds = fraud_df.frauds.str.replace("’","").str.replace("‘","")
```

```python
fraud_df.frauds.to_list()
```

```python
fraud_df.frauds = fraud_df.frauds.str.lower()
```

```python
fraud_df.frauds = fraud_df.frauds.str.split()
fraud_df.frauds
```

```python
sw = set(stopwords.words('english'))
punctuations = list(string.punctuation)
sw.update(punctuations)
```

```python
sp_char='[@_!#$%^&*()<>?/\|}{~£+:.-]™"'
lemmatizer = WordNetLemmatizer()
def clean_data(sentences):
    clean_sentences=[]
    for sentence in sentences:
        clean_sentence = []
        for word in sentence:
            w = word.strip()
            if (w not in sw) and (w not in sp_char) and w:
                clean_word = w
                clean_sentence.append(clean_word)
        clean_sentences.append(clean_sentence)
    return clean_sentences
```

```python
fraud_cleaned = clean_data(fraud_df.frauds.to_list())
fraud_cleaned
```

```python
cleaned_data=[]
for string in fraud_cleaned:
    cleaned_data.append(','.join(string))
```

```python
df= pd.DataFrame({'frauds':cleaned_data})
```

```python
text_data=df['frauds'].str.split(',',expand=True)
```
